## Pulling market value from transfer market website

```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
#list of player names
player_list = df['Player'].unique()
```

```python
# Base URL for player search
base_url = "https://www.transfermarkt.com/schnellsuche/ergebnis/schnellsuche?query="
```

```python
# DataFrame to store results
results = []

# Total number of players
total_players = len(player_list)
```

```python
# Iterate through the player list with a counter
# enumerate() → Gives both the index (i) and the player name.
# start=1 → Makes the counter start from 1 instead of 0.
for i, player in enumerate(player_list, start=1):
    # Format the player name for the URL
    search_url = base_url + player.replace(" ", "+")
    print(f"Processing player {i}/{total_players}: {player}")

    try:
        # Fetch the page
        # Sends a GET request with a browser-like user agent (to avoid blocking).
        # raise_for_status() → Throws an error if the request fails (e.g., 404).
        response = requests.get(search_url, headers={"User-Agent": "Mozilla/5.0"})
        response.raise_for_status()

        # Converts the HTML response into a BeautifulSoup object for easy data extraction.
        soup = BeautifulSoup(response.text, "html.parser")

        # Extract the market value
        # Looks for a <td> tag with the specific CSS classes that contain the market value.
        # If found, extracts the text; otherwise sets "Not Found".
        market_value_tag = soup.find("td", class_="rechts hauptlink")  # Target the specific class
        if market_value_tag:
            market_value = market_value_tag.text.strip()
        else:
            market_value = "Not Found"

        # Append the result
        results.append({"Player": player, "Market Value": market_value})
    except Exception as e:
        results.append({"Player": player, "Market Value": "Error: " + str(e)})

    # Waits 2 seconds between requests to avoid being blocked for too many requests in a short time.
    time.sleep(2)
```

```python
df_result=pd.DataFrame(results)
df_result
```

```python
# saving the dataframe
df_results.to_csv("player_market_values.csv", index=False)
```

### In summary:

- What it does: Goes through a list of players, searches their page, and extracts their market value.

- Why the delay: Prevents the website from blocking the scraper.

- Error handling: Ensures the script continues even if one player’s data can’t be fetched.
